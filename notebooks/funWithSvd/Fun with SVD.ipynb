```python
# the original image is here: http://www.flickr.com/photos/scequalizationschools/6029653081/
img = Image.open("14columbus_street_elementary_charleston.jpg")
img.mode
```

Loading the image gray scal pixels into a numpy array..

```python
# (If this fails, have a look at this post to install jpeg support: 
# http://stackoverflow.com/questions/4435016/install-pil-on-virtualenv-with-libjpeg)
pixels = img.getdata()

original = np.array(list(pixels)).reshape(img.size[1],img.size[0] )

# we want the zero-mean version, otherwise X.T dot X is not the scaled correlation matrix 
original = original - np.mean(original)

pixSize = img.size[0] * img.size[1]
pixSize, original.shape
```

```python
%%R 
    # Yeah! They enabled R code in iPython notebooks! If that's not cool, what is? :-)

    # display the grey scale levels, like a heatmap
    showImage <- function(img) {
        x <- seq(1, dim(img)[2])
        y <- seq(dim(img)[1], 1, -1)
        img.df <- expand.grid(x=x, y=y)
        img.df$pixval <- as.vector(t(img))
    
        g <- ggplot(img.df, aes(x=x, y=y)) + 
             geom_tile(aes(fill=pixval)) +
             scale_fill_gradient(low="black", high="white")
        print (g)
    }
```

```python
# let's have a look at the original image... 
%R -i original -w 600 -h 380 showImage(original)
```

## Compressing the image with PCA

### Building the model

```python
# SVD:
U, S, Vt = np.linalg.svd(original.T)

# PCA:
index = np.argsort(S)[::-1]
V = Vt.T
U = U[:,index]
S = S[index]
V = V[:,index]
```

### Looking at the model

```python
# let's look at the top singular values
varExp = S / sum(S)
```

```python
%%R -i varExp
    plotHisto <- function(vals) {
        df <- data.frame(x=seq(1,length(vals)), y=vals)
        g <- ggplot(df) + 
               geom_bar(aes(x=x, y=y), stat="identity") +
               xlab("sing. vector index") + 
               ylab("percentage of explained variance")
        print (g)
    }

    plotHisto(varExp)
```

```python
# zoom on the first ones
firstOnes = varExp[:20]
%R -i firstOnes plotHisto(firstOnes)
```

```python

for n in [5, 26, 45, 74, 119, 193]:
    print "%d vectors account for %.2f %% of the image variance" % (n, 100*sum(varExp[:n]))
```

### Compression methods

```python
# (usage of transpose below because my image is horizontal...)

covEigBase = V

# image compression, using the 'size' first column of the eigen base of the correlation matrix
compress = lambda image, size : np.dot(image.T, covEigBase[:,:size])

    
# decompression, just the reverse operation
decompress = lambda compressed, size : np.dot(compressed, covEigBase[:,:size].T).T


def compressAndDecompress(image, size): 
    compressed = compress (original, size)
    decompressed = decompress(compressed, size)
    return compressed, decompressed
```

### A few compression results

```python
compressedFull, decompressedFull = compressAndDecompress (original, 394)
compressedFull.shape
```

```python
%R -i decompressedFull -w 600 -h 380 showImage(decompressedFull)
```


Right, so using the full eigen base yields a "compressed" image with the same size as the original and a decompressed one with no quality loss. 

So far nothing surprising...
